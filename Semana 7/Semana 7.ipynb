# Día 1

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

datos = {
    "pais": ["España", "Francia", "Italia", "Alemania", "Portugal", "Grecia"],
    "PIB": [1400, 2800, 2100, 3800, 1300, 1200],
    "inflacion": [2.5, 2.1, 2.8, 1.9, 3.0, 3.5],
    "paro": [12, 8, 10, 5, 11, 16]
}

df = pd.DataFrame(datos)

df
```

# Día 2
## ¿PIB y paro están relacionados?

```python
#scatter PIB vs paro
plt.scatter(df["PIB"], df["paro"])

#correlación PIB vs paro
print("Correlación:", df["PIB"].corr(df["paro"]))

#regresión lineal simple
m, b = np.polyfit(df["PIB"], df["paro"], 1)
plt.plot(df["PIB"], m*df["PIB"] + b)
print("Pendiente:", m)

#etiqutas gráfica
plt.title("Regresión PIB vs Paro")
plt.xlabel("PIB")
plt.ylabel("Paro")
plt.show()
```

## El PIB y el paro tienen una correlación negativa fuerte por lo que están relacionados inversamente, cuando el PIB sube el paro baja.

# Día 3
## ¿Inflación y paro están relacionados?

```python
#scatter inflación vs paro
plt.scatter(df["inflacion"], df["paro"])

#correlación inflación vs paro
print("Correlación:", df["inflacion"].corr(df["paro"]))

#regresión lineal simple
m, b = np.polyfit(df["inflacion"], df["paro"], 1)
plt.plot(df["inflacion"], m*df["inflacion"] + b)

print("Pendiente:", m)

#etiqutas gráfica
plt.title("Regresión Inflación vs Paro")
plt.xlabel("Inflación")
plt.ylabel("Paro")
plt.show()
```

## El paro y la inflación tienen una correlación positiva fuerte, por lo que no corresponde con lo que afirma la curva de Philips, esto pude deberse a que la muestra es artificial. Las dos variables tienen una relación directa.

# Día 4
## ¿PIB e inflación están relacionados?

```python
#scatter PIB vs inflación
plt.scatter(df["PIB"], df["inflacion"])

#correlación PIB vs inflación
print("Correlación:", df["PIB"].corr(df["inflacion"]))

#regresión lineal simple
m, b = np.polyfit(df["PIB"], df["inflacion"], 1)
plt.plot(df["PIB"], m*df["PIB"] + b)

print("Pendiente:", m)

#etiqutas gráfica
plt.title("Regresión PIB vs Inflación")
plt.xlabel("PIB")
plt.ylabel("Inflación")
plt.show()
```

## El PIB y la inflación tienen una relación inversa, cuando el PIB sube la inflación baja. El coeficiente de correlación de Pearson es coherente con la RLS porque las variables mantienen una correlación negativa fuerte (-0.85).
