```python
import pandas as pd
import matplotlib.pyplot as plt

datos = { 
    "pais": ["España", "Francia", "Italia", "Alemania"],
    "PIB": [1400, 2800, 2100, 3800],
    "inflacion": [2.5, 2.1, 2.8, 1.9]
}

df = pd.DataFrame(datos)

plt.bar(df["pais"], df["PIB"])
plt.show()
```

```python
plt.bar(df["pais"], df["inflacion"])
plt.show()
```

```python
plt.scatter(df["PIB"], df["inflacion"])
plt.show()
```
