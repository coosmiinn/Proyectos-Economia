```python
import pandas as pd
import matplotlib.pyplot as plt

datos = { 
    "pais": ["Japon", "Corea del Sur", "EEUU"],
    "PIB": [3500, 4000, 2500],
    "inflacion": [3.4, 1.2, 3.1]
}

df = pd.DataFrame(datos)

plt.bar(df["pais"], df["PIB"])
plt.title("PIB por pais")
plt.xlabel("Pais")
plt.ylabel("PIB")
plt.show()
```

```python
plt.scatter(df["PIB"], df["inflacion"])
plt.show()
```
