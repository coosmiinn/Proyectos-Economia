```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

datos = {
    "pais": ["España", "Francia", "Italia", "Alemania", "Portugal", "Grecia"],
    "PIB": [1400, 2800, 2100, 3800, 1300, 1200],
    "inflacion": [2.5, 2.1, 2.8, 1.9, 3.0, 3.5],
}

df = pd.DataFrame(datos)

# correlación
print("Coeficiente de correlación de Pearson:", df["PIB"].corr(df["inflacion"]))

# scatter PIB vs inflación
plt.scatter(df["PIB"], df["inflacion"])

# regresión lineal simple
m, b = np.polyfit(df["PIB"], df["inflacion"], 1)
plt.plot(df["PIB"], m*df["PIB"] + b)
print("Pendiente:", m)

# etiquetas gráfica
plt.xlabel("PIB")
plt.ylabel("Inflación")
plt.title("PIB vs Inflación")

plt.show()
```

## Se puede observar en el resultado obtenido a partir del coeficiente de correlación de Pearson que existe una correlación negativa fuerte entre el PIB y la inflación de estos 6 países. Para hacerlo más visual se ha creado un gráfico de dispersión y se le ha aplicado una regresión lineal simple que ha resultado tener una pendiente negativa lo que confirma el resultado del coeficiente de correlación de Pearson y también confirma que las dos variables tienen una relación inversa, cuanfo el PIB sube la inflación baja.
