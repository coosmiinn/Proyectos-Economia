```python
import pandas as pd
import matplotlib.pyplot as plt 

datos = {
    "PIB": [1400, 2800, 2100, 3800],
    "inflacion": [2.5, 2.1, 2.8, 1.9]
}

df = pd.DataFrame(datos)

plt.scatter(df["PIB"], df["inflacion"])
plt.xlabel("PIB")
plt.ylabel("Inflación")
plt.title("Relación PIB vs Inflación")
plt.show()

print(df["PIB"].corr(df["inflacion"]))
```

## Según el gráfico obtenido al comparar el PIB con la inflación se puede ver que las dos variables tienen una correlación negativa fuerte si trazasemos una línea alrededor de la nube de puntos. Esta conclusión también nos la confirma el resultado que se ha obtenido utilizando el coeficiente de correlación de Pearson (-0.81).
