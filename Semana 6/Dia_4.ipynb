```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

datos = {
    "PIB": [1400, 2800, 2100, 3800],
    "inflacion": [2.5, 2.1, 2.8, 1.9] 
}

df = pd.DataFrame(datos)

#scatter
plt.scatter(df["PIB"], df["inflacion"])

#línea de tendencia
m, b = np.polyfit(df["PIB"], df["inflacion"], 1)
plt.plot(df["PIB"], m*df["PIB"] + b)

plt.show()
```

## En el gráfico que se muestra se puede observar que se ha aplicado una regresión lineal simple y se ha obtenido una pendiente negativa lo que significa que el PIB y la inflación tienen una correlación negativa. No obstante, dicho resultado hay que tomarselo con cautela al ser obtenido mediante una fuente de datos escaza.
