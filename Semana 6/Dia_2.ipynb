```python
import pandas as pd

datos = {
    "PIB": [1400, 2800, 2100, 3800],
    "inflacion": [2.5, 2.1, 2.8, 1.9],
    "paro": [12, 8, 10, 5]
}

df = pd.DataFrame(datos)

print(df.corr())
```

## Estas correlaciones son de tipo lineal (coeficiente de Pearson). El PIB y la inflación tienen una correlación negativa furte (-0.8). El PIB y el paro tienen una correlación negativa practicamente perfecta (-0.99) que es coherente con la ley de Okun. El paro y la inflación tienen una correlación positiva fuerte (0.8), aunque según la curva de Phillips tradicional debería ser negativa. Importante es no tomarse estos resultados demasiado en serio ya que la muestra es muy reducida.
